```python
import networkx as nx
import matplotlib.pyplot as plt

# Nodes
members = [
    "Dr. Ackah",
    "Bismark",
    "Naana",
    "Joseph",
    "Mike",
    "Priscilla",
    "Fred"
]

# Edges (relationships)
relationships = [
    ("Dr. Ackah", "Bismark"),
    ("Dr. Ackah", "Naana"),
    ("Dr. Ackah", "Joseph"),
    ("Dr. Ackah", "Mike"),
    ("Dr. Ackah", "Priscilla"),
    ("Dr. Ackah", "Fred"),
    ("Bismark", "Naana"),
    ("Bismark", "Joseph"),
    ("Naana", "Joseph"),
    ("Mike", "Priscilla"),
    ("Mike", "Fred"),
    ("Priscilla", "Fred"),
]

G = nx.Graph()
G.add_nodes_from(members)
G.add_edges_from(relationships)

print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
print("Degree distribution:", dict(G.degree()))
print("Isolated nodes:", list(nx.isolates(G)))

# Draw network
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed=42)
nx.draw(
    G, pos,
    with_labels=True,
    node_size=1000,
    node_color="skyblue",
    edge_color="gray",
    font_size=10
)
plt.title("Research Group Network")
plt.tight_layout()
plt.show()

# Export for Gephi
nx.write_gexf(G, "research_group_network.gexf")
```
